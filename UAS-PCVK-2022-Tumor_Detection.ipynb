```python
from IPython.core.pylabtools import figsize
index = 0
for image_ori in filename:
  img = cv2.imread(image_ori) 
  img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  blur = cv2. GaussianBlur (img, (5,5),0)
      
  ret, th1 = cv2.threshold(blur, 165, 255, cv2.THRESH_BINARY)
  kernel = np.ones((7,7), np.uint8)
  opening = cv2.morphologyEx(th1, cv2.MORPH_OPEN, kernel)

  
  thresh = cv2.Canny(opening,40,185)
  contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
  shaped = cv2.drawContours(img, contours, -1, (255,0,0), 3)
  

  titles = ['Real Image', 'Global Thresholding (v = 170)','Opening Morphology', 'shaped'] 
  citra3 = [blur, th1, opening, shaped]

  plt. figure(figsize = (20,15)) 
  
  for i in range(len(citra3)):
      plt. subplot(1,4,i+1),plt.imshow(citra3[i])
      
      plt.title(titles[i])
      plt.xticks([]),plt.yticks([]) 
  plt.show()


  
```
